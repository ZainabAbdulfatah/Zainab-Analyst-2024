```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud


# Load dataset
df = pd.read_excel("/content/Naija_PeaceTech_Survey_-_all_versions_-_English_en_-_2025-09-05-20-18-23.xlsx")
print("Shape of dataset:", df.shape)
print(df.head())
print("\nOriginal DataFrame columns:\n", df.columns.tolist())
```

```python
# Print all columns to inspect actual names
for i, col in enumerate(df.columns, 1):
    print(f"{i}. {repr(col)}")
```

```python
# Clean and normalize column names
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(r"\s+", "_", regex=True)
df.columns = df.columns.str.replace(r"[^\w]", "", regex=True)
df.columns = df.columns.str.lower()

print("\nCleaned column names:\n", df.columns.tolist())
```

```python
drop_cols = [
    "_id", "_uuid", "_submission_time", "_validation_status", "_notes",
    "_status", "_submitted_by", "__version__", "_tags", "_index"
]

df = df.drop(columns=[c for c in drop_cols if c in df.columns])
print("\nShape after dropping system columns:", df.shape)

# === key demographic columns ===
if "q1_whatisyourage" in df.columns:
    print("\nAge distribution:\n", df["q1_whatisyourage"].value_counts())

if "q2_whatisyourgender" in df.columns:
    print("\nGender distribution:\n", df["q2_whatisyourgender"].value_counts())

if "what_state_do_you_currently_reside_in" in df.columns:
    print("\nTop 10 States:\n", df["what_state_do_you_currently_reside_in"].value_counts().head(10))

if "what_lga_do_you_currently_reside_in" in df.columns:
    print("\nTop 10 LGAs:\n", df["what_lga_do_you_currently_reside_in"].value_counts().head(10))
```

```python
# Check missing values
missing = df.isnull().sum().sort_values(ascending=False)
print("Missing values:\n", missing.head(20))

# Drop irrelevant metadata columns
meta_cols = [c for c in df.columns if any(x in c.lower() for x in ["start", "end", "device", "id", "submit"])]

# Explicitly exclude the gender column from meta_cols
gender_col_to_keep = "Q2. What is your gender"
if gender_col_to_keep in meta_cols:
    meta_cols.remove(gender_col_to_keep)


df_clean = df.drop(columns=meta_cols)

# Handle duplicates
df_clean = df_clean.drop_duplicates()

print("New shape after cleaning:", df_clean.shape)
```

```python
drop_cols = [
    "Katherine Tyochir ",
    "Email (optional)",
    "Enumerator Name",
    "Q21. If yes, please provide your contact (Whatsapp Preferred, Kept Confidential)",
    "What_state_and_LGA_d_currently_reside_in",
    "_validation_status", "_notes", "_tags",
    "_submitted_by"
]

# Drop only if the column exists
df_clean = df_clean.drop(columns=[c for c in drop_cols if c in df_clean.columns])

print("Shape after dropping irrelevant:", df_clean.shape)
```

```python
# Descriptive stats for numerical fields
print(df_clean.describe())

# Value counts for categorical fields (Top 5 per column)
for col in df_clean.select_dtypes(include="object").columns:
    print(f"\n{col}\n", df_clean[col].value_counts().head())
```

```python
# clean multi-select responses into binary (1/0)
def clean_multiselect(df, prefix):
    q_cols = [c for c in df.columns if c.lower().startswith(prefix)]
    if not q_cols:
        print(f"No columns found for {prefix.upper()}")
        return None

    # Convert Yes/No or non-null into 1/0
    df_binary = df[q_cols].notna().astype(int)

    counts = df_binary.sum().sort_values(ascending=False)
    print(f"\n--- {prefix.upper()} Summary ---")
    print(counts)

    return counts

# Apply to multi-select questions
q9_counts  = clean_multiselect(df_clean, "q9")
q13_counts = clean_multiselect(df_clean, "q13")
q14_counts = clean_multiselect(df_clean, "q14")
q16_counts = clean_multiselect(df_clean, "q16")
```

```python
def plot_gender_pie_chart(df, title="What is Your Gender", exclude=["Other", ""]):
    column = "q2_what_is_your_gender"  # exact column name

    # Ensure column exists
    if column not in df.columns:
        print(f"Column '{column}' not found in dataframe.")
        return

    # Filter unwanted values
    filtered_df = df[~df[column].isin(exclude)]
    if filtered_df.empty:
        print(f"No data to plot for {title}")
        return

    # Count occurrences
    counts = filtered_df[column].value_counts()

    # Pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(
        counts.values,
        labels=counts.index,
        autopct="%1.1f%%",
        startangle=140,
        colors=plt.cm.Paired.colors
    )
    plt.title(title, fontsize=16)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_categorical_distribution(df, column, title="Distribution", exclude=["Other", ""], palette="viridis"):
    # Ensure column exists
    if column not in df.columns:
        print(f"Column '{column}' not found in dataframe.")
        return

    # Filter unwanted values
    filtered_df = df[~df[column].isin(exclude)]
    if filtered_df.empty:
        print(f"No data to plot for {title}")
        return

    # Count occurrences
    counts = filtered_df[column].value_counts()
    percentages = counts / counts.sum() * 100

    # Plot
    plt.figure(figsize=(8, 6))
    sns.barplot(
        x=counts.index,
        y=counts.values,
        palette=sns.color_palette(palette, n_colors=len(counts))
    )

    plt.title(title, fontsize=16)
    plt.xlabel(column.replace("_", " ").title(), fontsize=12)
    plt.ylabel("Count", fontsize=12)

    # Add count + percentage labels
    for index, value in enumerate(counts.values):
        plt.text(
            index,
            value + max(counts.values) * 0.01,  # slight offset
            f"{value} ({percentages.values[index]:.1f}%)",
            ha="center",
            fontsize=10
        )

    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

edu_col = "what_is_your_highest_level_of_education_completed"

# Step 1: Drop blanks and NaN
edu_data = df[edu_col].dropna().astype(str).str.strip()

# Step 2: Remove anything that looks like "Other" or "Option"
edu_data = edu_data[~edu_data.str.contains("other", case=False, na=False)]
edu_data = edu_data[~edu_data.str.contains("option", case=False, na=False)]

# Step 3: Count frequencies
edu_counts = edu_data.value_counts().to_dict()

# Step 4: Generate WordCloud
wordcloud = WordCloud(
    width=800, height=400,
    background_color="white",
    colormap="viridis"
).generate_from_frequencies(edu_counts)

# Step 5: Plot
plt.figure(figsize=(10,6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Education Level Distribution (Cleaned)", fontsize=14, weight="bold")
plt.show()
```

```python
def plot_categorical_distribution(
    df,
    column,
    title="Distribution",
    exclude=["Other", ""],
    top_n=None
):
    # Ensure column exists
    if column not in df.columns:
        print(f"Column '{column}' not found in dataframe.")
        return

    # Filter unwanted values
    filtered_df = df[~df[column].isin(exclude)]

    if filtered_df.empty:
        print(f"No data to plot for {title}")
        return

    # Count occurrences
    counts = filtered_df[column].value_counts()

    if top_n:
        counts = counts.head(top_n)  # Take only top_n values

    # Wider figure for long labels
    plt.figure(figsize=(14, 8))

    # Normal vertical bar chart without palette warning
    sns.barplot(
        x=counts.index,
        y=counts.values,
        palette="Blues_d",
        hue=None,
        dodge=False
    )

    plt.title(title, fontsize=16)
    plt.xlabel(column.replace("_", " ").title(), fontsize=12)
    plt.ylabel("Count", fontsize=12)

    # Add count labels
    for i, value in enumerate(counts.values):
        plt.text(i, value, str(value), ha="center", va="bottom", fontsize=10)

    # Rotate labels for readability
    plt.xticks(rotation=45, ha="right", fontsize=10)

    plt.tight_layout()
    plt.show()
```

```python

# Word Cloud for states
def plot_state_wordcloud(df, column, title):
    text = " ".join(str(state) for state in df[column].dropna())
    wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)
    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title(title, fontsize=16)
    plt.show()

# Bar chart for top N LGAs
def plot_local_government_distribution(df, column, title, top_n=15):
    data_counts = df[column].value_counts().nlargest(top_n)
    plt.figure(figsize=(12,6))
    sns.barplot(y=data_counts.index, x=data_counts.values, palette="viridis")
    plt.title(title, fontsize=16)
    plt.xlabel("Count")
    plt.ylabel("Local Government")
    plt.show()

# Usage
plot_state_wordcloud(df, "what_state_do_you_currently_reside_in", "State of Residence Distribution")
plot_local_government_distribution(df, "what_lga_do_you_currently_reside_in", "Local Government Distribution", top_n=15)
```
